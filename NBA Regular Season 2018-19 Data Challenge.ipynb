# Nba Regular Season 2018-19 Data Challenge

Your task will be to take the dataset given, and create an analysis answering the following 10 questions. This project will again test your knowledge of pandas in order to find the answers needed given the data you are presented with.

```python
import numpy as np
import pandas as pd
```

# What was the average age of player in the league?

```python
# Setting all DataFrames displayed to show all columns
pd.set_option('display.max_columns', None)

nba2018 = pd.read_csv('nbastats2018-2019.csv', sep = ',')
nba2018
```

```python
# Setting display to show numbers in data frames to just 2 decimal places
pd.options.display.float_format = '{:,.2f}'.format

age = nba2018[['Age']].mean(numeric_only = True)
print("Average", age)
```

# What player scored the most points?

```python
pd.options.display.float_format = '{:,.0f}'.format
nba2018['Total Points'] = nba2018['Points']*nba2018['G']

most_points = nba2018[['Name', 'Total Points']].sort_values('Total Points', ascending = False)
most_points.head()
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
nba2018['Total Blocks'] = nba2018['Blocks']*nba2018['G']
nba2018['Total Steals'] = nba2018['Steals']*nba2018['G']
nba2018['Total Assists'] = nba2018['Assists']*nba2018['G']
nba2018['Total Rebounds'] = nba2018['Rebounds']*nba2018['G']

most_blocks = nba2018[['Name', 'Total Blocks']].sort_values('Total Blocks', ascending = False)
most_blocks.head()

# Myles Turner is a Center on the Indiana Pacers
```

# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
win_shares = nba2018[['Name', 'Team', 'WS']].sort_values('WS', ascending = False)
win_shares.head()
```

# What player had the best 3-pt percentage? 

```python
pd.options.display.float_format = '{:,.2f}'.format
three_pointers = nba2018[nba2018['FG3A'] >= 3.0]
best_shooter = three_pointers[['Name', 'FG3%', 'FG3A']].sort_values('FG3%', ascending = False)
best_shooter['FG3%'] = best_shooter['FG3%'].apply(lambda x: round(x, 3))
best_shooter.head()

# Based on the threshold of attempting at least 3 three pointers per game, Joe Harris had the best 3-pt percentage at 47.4%
```

# Who played the most minutes during the season

```python
pd.options.display.float_format = '{:,.0f}'.format
nba2018['Total Minutes'] = nba2018['MP']*nba2018['G']

most_minutes = nba2018[['Name', 'Total Minutes']].sort_values('Total Minutes', ascending = False)
most_minutes.head()
```

# What player given their player effiecency rating was the clutchest during the season?

```python
pd.options.display.float_format = '{:,.2f}'.format
per = nba2018[nba2018['MP'] >= 20.0]
most_efficient = per[['Name', 'PER']].sort_values('PER', ascending = False)
most_efficient.head()

# Based on the threshold of a player needing to play at least 20 minutes per game, Giannis Antetokounmpo was the most clutch
```

# What team had the youngest roster?

```python
youngest = nba2018.groupby('Team').mean(numeric_only = True)[['Age']].sort_values('Age', ascending = True)
youngest.head()
```

# Who is the highest paid player during the season?

```python
rich = nba2018.loc[nba2018['Salary'] != "-"]
rich['Salary'] = pd.to_numeric(rich['Salary'])
richest = rich[['Name', 'Salary']].sort_values('Salary', ascending = False)
richest.loc[:, 'Salary'] = richest['Salary'].map('{:,d}'.format)
richest.head()
```
